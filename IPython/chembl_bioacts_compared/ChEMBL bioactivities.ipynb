## Bioactivities
Download all bioactivities for targets

```python
def fetchBioacts(tgtsList):
    """
    Fetches bioactivities and returns dataframe
    Expects list of targets
    """
    bioactsDF = ""
    for k in tgtsList:
        t = targets.bioactivities(k)
        tmpdf = pd.DataFrame(t)
        if len(bioactsDF) > 0:
            bioactsDF = bioactsDF.append(tmpdf)
        else:
            bioactsDF = tmpdf
    bioactsDF.index = range(len(bioactsDF)) # reindex
    return bioactsDF
```

```python
bioactsDF = fetchBioacts(list(targetsDF['chemblId']))
```

```python
bioactsDF.head(1)
```

```python
len(bioactsDF)
```

Filter bioactivities (safe defaults)

```python
bioactsDF = bioactsDF[(bioactsDF['bioactivity_type'] == 'IC50') &  # keep ony IC50
                      (bioactsDF['operator'] == '=') & # only exact measurements
                      (bioactsDF['assay_type'] == 'B') & # only binding data
                      (bioactsDF['target_confidence'] >= 8)] # only high target confidence
```

```python
bioactsDF = bioactsDF[bioactsDF['value'].notnull()]
bioactsDF = bioactsDF[bioactsDF['value'] != 'Unspecified']
```

```python
bioactsDF['value'] = bioactsDF['value'].astype(float) # making sure everything is float
```

#### Add bioactivities to cpdsDF

```python
def getBioacts(cpd, target, targetConfidence):
    value = bioactsDF[(bioactsDF['ingredient_cmpd_chemblid'] == cpd)& # All rows of compound AND
                     (bioactsDF['target_chemblid'] == target)&
                     (bioactsDF['target_confidence'] == targetConfidence)]['value'].median() # target. Get mean of values
    return value # returns IC50     
```

Helper functions that count values, actives, calculate actives/all values

```python
def howmany(x):
    # Count all non-null values
    return x.count()
def belowThreshold(x, threshold):
    # all values below threshold
    return len(x[x <= threshold])
def fraction(x, threshold):
    # ratio between actives and all assay results
    return len(x[x <= threshold])/float(x.count())
```

```python
summaryDF = pd.DataFrame(columns=cpdsDF.columns)
```

```python
summaryDF = summaryDF.append(cpdsDF.apply(lambda x: howmany(x) ,axis=0), ignore_index=True)
summaryDF = summaryDF.append(cpdsDF.apply(lambda x: belowThreshold(x, 500) ,axis=0), ignore_index=True)
summaryDF = summaryDF.append(cpdsDF.apply(lambda x: fraction(x, 500) ,axis=0), ignore_index=True)
```

```python
summaryDF = summaryDF.drop(['chemblId','smiles'], axis=1)
summaryDF.index = ['Total','Active','Fraction of actives']
```
